**L1 and L2 Regualizatonin Neural Network**

Why Regularization?
 - Regularization helps prevent overfitting.


**What is L1 and L2 Regularization?**
* L1 - Formula: λ * Σ|wi|
* L2 - Formula: λ * Σ(wi^2)

image.png

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l1, l2
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# For binary classification, we'll select only two classes (setosa and versicolor)
X = X[y != 2]  # exclude the virginica class
y = y[y != 2]  # binary labels (0 = setosa, 1 = versicolor)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features (important for regularization)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model_l2 = Sequential([
    Dense(64, input_dim=4, activation='relu', kernel_regularizer = l2(0.01)),
    Dense(32, activation='relu', kernel_regularizer=l2(0.01)),
    Dense(1,activation='sigmoid')
])

model_l2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_l2.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
model_l1 = Sequential([
    Dense(64, input_dim=4, activation='relu', kernel_regularizer = l1(0.01)),
    Dense(32, activation='relu', kernel_regularizer=l1(0.01)),
    Dense(1,activation='sigmoid')
])

model_l1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_l1.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
# Get weights from the first layer of the models
weights_l2 = model_l2.layers[0].get_weights()[0]
weights_l1 = model_l1.layers[0].get_weights()[0]
```

```python

# Plotting the histograms of the weights
plt.figure(figsize=(14, 6))

# L2 regularization weight histogram
plt.subplot(1, 2, 1)
plt.hist(weights_l2.flatten(), bins=30, color='blue', alpha=0.7)
plt.title("L2 Regularization Weights Distribution")
plt.xlabel("Weight Value")
plt.ylabel("Frequency")

# L1 regularization weight histogram
plt.subplot(1, 2, 2)
plt.hist(weights_l1.flatten(), bins=30, color='red', alpha=0.7)
plt.title("L1 Regularization Weights Distribution")
plt.xlabel("Weight Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
